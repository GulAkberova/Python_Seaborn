```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

Bar plot

```python
tips = sns.load_dataset('tips')
```

```python
sns.barplot(x='day',y='tip',data=tips)
```

```python
sns.barplot(x='day',y='total_bill', data=tips)
```

```python
sns.barplot(x='day',y='total_bill', data=tips, hue='sex')
```

```python
sns.barplot(x='day',y='total_bill', data=tips, hue='sex',palette='winter_r')
```

```python
sns.barplot(x='day',y='total_bill', data=tips, hue='smoker')
```

```python
sns.barplot(x='total_bill',y='day', data=tips, palette="spring")
```

```python
sns.barplot(x='day', y='tip', data=tips, palette='spring', order=['Sat','Fri', 'Sun', 'Thur'])
```

```python
from numpy import median
```

```python
 sns.barplot(x='day',y='total_bill', data=tips, estimator=median,palette='spring')
```

```python
sns.barplot(x='smoker', y='tip', data=tips, estimator=median, hue='sex', palette='coolwarm')
```

```python
sns.barplot(x='smoker',y='tip', data=tips, ci=99)
```

```python
sns.barplot(x='smoker',y='tip', data=tips, ci=58,palette='winter_r',estimator=median)
```

```python
sns.barplot(x='day', y='total_bill', data=tips, palette='spring',capsize=0.3)
```
