```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
iris= sns.load_dataset('iris')
```

```python
iris.head()
```

```python
x=sns.PairGrid(iris)
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris)
x=x.map_diag(plt.hist)
```

```python
x=sns.PairGrid(iris)
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue='species')
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris, hue='species')
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue='species',palette='coolwarm') #husl,winter_r,RdBu
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue='species',palette='autumn')
x=x.map_diag(plt.hist,histtype='step')
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue='species',palette='coolwarm')
x=x.map_diag(plt.hist,histtype='step',linewidth=3)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris)
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris, vars=['petal_length','petal_width'])
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue='species', vars=['petal_length','petal_width'],palette='husl')
x=x.map_diag(plt.hist,edgecolor='black')
x=x.map_offdiag(plt.scatter,edgecolor='black')
x=x.add_legend()
```

```python
x=sns.PairGrid(iris, x_vars=['petal_width','petal_length'],
              y_vars=['sepal_width','sepal_width'])
x=x.map(plt.scatter,color='green',edgecolor='black')
```

```python
x=sns.PairGrid(iris,palette='coolwarm',hue='species')
x=x.map_diag(plt.hist)
x=x.map_upper(plt.scatter)
x=x.map_lower(sns.kdeplot)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue='species',palette='winter_r',hue_kws={'marker':['D','s','+']})
x=x.map(plt.scatter, s=30)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue='species',palette='RdBu',hue_kws={'marker':['D','s','D']})
x=x.map(plt.scatter, s=30, edgecolor='gray')
x=x.add_legend()
```
