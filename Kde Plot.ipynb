Kernel Dens Estim

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
mean=[0,0]
cov=[[0.2,0],[0,3]]
```

```python
x_axis, y_axis = np.random.multivariate_normal(mean, cov, size=40).T
```

```python
sns.kdeplot(x_axis)
```

```python
sns.kdeplot(x_axis,shade=True)
```

```python
sns.kdeplot(x_axis,shade=True,color='green')
```

```python
sns.kdeplot(y_axis,shade=True, color='purple')
```

```python
sns.kdeplot(x=x_axis , y=y_axis)
```

```python
sns.kdeplot(x=x_axis , y=y_axis, shade=True)
```

```python
sns.kdeplot(x=x_axis , y=y_axis, cmap='RdBu')
```

```python
sns.kdeplot(x=x_axis , y=y_axis,n_levels=68, cmap='husl')
```

```python
sns.kdeplot(y_axis, bw=2.5)
```

```python
sns.kdeplot(y_axis, bw=0.1)
```

```python
sns.kdeplot(y_axis,vertical=True)
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
setosa=iris.loc[iris.species =='setosa']
versicolor=iris.loc[iris.species == 'versicolor']
```

```python
sns.kdeplot(x=setosa.petal_length,y=setosa.petal_width,cmap='coolwarm')
sns.kdeplot(x=versicolor.petal_length,y=versicolor.petal_width, cmap='RdBu')
```

```python
sns.kdeplot(x=setosa.petal_length,y=setosa.petal_width,cmap='coolwarm',shade=True)
sns.kdeplot(x=versicolor.petal_length,y=versicolor.petal_width, cmap='RdBu',shade=True)
```

```python
sns.kdeplot(x=setosa.petal_width,y=setosa.petal_length,cmap='coolwarm', n_levels=35)
sns.kdeplot(x=versicolor.petal_width,y=versicolor.petal_length, cmap='RdBu',n_levels=35)
```
