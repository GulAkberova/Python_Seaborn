```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
sns.FacetGrid(tips,row='smoker',col='time')
```

```python
x=sns.FacetGrid(tips,row='smoker',col='time')
x=x.map(plt.hist,'total_bill')
```

```python
x=sns.FacetGrid(tips,row='smoker',col='time')
x=x.map(plt.hist,'total_bill',color='green')
```

```python
x=sns.FacetGrid(tips,row='smoker',col='time')
x=x.map(plt.hist,'total_bill',color='blue',bins=15)
```

```python
x=sns.FacetGrid(tips,col='time',row='smoker')
x=x.map(plt.scatter,'total_bill','tip')
```

```python
x=sns.FacetGrid(tips,col='time',row='smoker')
x=x.map(sns.regplot,'total_bill','tip')
plt.show()
```

```python
x=sns.FacetGrid(tips,col='time',row='smoker')
x=x.map(plt.scatter,'total_bill','tip')
```

```python
x=sns.FacetGrid(tips,col='time',hue='smoker')
x=x.map(plt.scatter,'total_bill','tip')
```

```python
x=sns.FacetGrid(tips,col='time',hue='smoker')
x=x.map(plt.scatter,'total_bill','tip')
x=x.add_legend()
```

```python
x=sns.FacetGrid(tips,col='day')
x=x.map(sns.boxplot,'total_bill','time')
```

```python
x=sns.FacetGrid(tips,col='day',height=5,aspect=.5)
x=x.map(sns.boxplot,'time','total_bill')
```

```python
x=sns.FacetGrid(tips,col='day',col_order=['Sat','Thur','Sun','Fri'], height=5,aspect=.5)
x=x.map(sns.boxplot,'time','total_bill',color='red')
```

```python
x=sns.FacetGrid(tips,col='time',hue='smoker',palette='winter_r',height=4,aspect=.7)
x=x.map(plt.scatter,'total_bill','tip')
x=x.add_legend()
```
