```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.violinplot(tips)
```

```python
sns.violinplot(x=tips['tip'])
```

```python
sns.violinplot(x='size',y='total_bill',data=tips,palette='winter_r')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',palette='spring')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',palette='coolwarm',dodge=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',palette='coolwarm',dodge=True,
               order=['Sat','Thur','Sun','Fri'],split=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',palette='autumn',
              inner='quartile')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',palette='autumn',
              inner='quartile',split=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',palette='RdBu',
              inner='quartile',split=False,scale='count')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',palette='winter_r',
              inner='stick',split=False,scale='count',scale_hue=False)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',palette='winter_r',
              inner='stick',split=False,scale='count',scale_hue=False,bw=0.1)
```
