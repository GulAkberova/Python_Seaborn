```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
normal=np.random.rand(12,15)
```

```python
sns.heatmap(normal)
```

```python
sns.heatmap(normal,annot=True)
```

```python
sns.heatmap(normal,vmin=0,vmax=2)
```

```python
sns.heatmap(normal,vmin=0,vmax=0.5)
```

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights.tail()
```

```python
flights = flights.pivot(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(flights)
```

```python
sns.heatmap(flights,annot=True)
```

```python
sns.heatmap(flights,annot=True,fmt='d')
```

```python
sns.heatmap(flights,linewidth=0.9,annot=True,fmt='d')
```

```python
sns.heatmap(flights,linewidth=0.9,annot=True,fmt='d',vmin=100,vmax=650,cmap='RdBu')
```

```python
sns.heatmap(flights,linewidth=0.9,annot=True,fmt='d',vmin=100,vmax=650,cmap='summer')
```

```python
sns.heatmap(flights,center=flights.loc['Jun',1954],cmap='summer',annot=True,fmt='d',
           cbar=False)
```
