```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.regplot(x='total_bill', y='tip', data=tips,color='red')
```

```python
mean=[2,5]
cov=[[1.1,0.4],[2.2,3]]
```

```python
x_value, y_value= np.random.multivariate_normal(mean, cov, 100).T

sns.regplot(x=x_value, y=y_value,color='green')
```

```python
series_x_value=pd.Series(x_value, name='Var_X')
series_y_value=pd.Series(y_value, name='Var_Y')

sns.regplot(x=series_x_value, y=series_y_value, marker='D',color='purple')
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,marker='D',
           scatter_kws={'color':'blue'},
           line_kws={'color':'red','linewidth':5.1})
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,marker='o',
           scatter_kws={'color':'purple'},
           line_kws={'color':'green','linewidth':1.1})
```

```python
sns.regplot(x=x_value, y=y_value, ci=84)
```

```python
sns.regplot(x='size', y='total_bill',data=tips, ci=84,color='green')
```

```python
sns.regplot(x='size', y='total_bill',data=tips, ci=84,color='purple',x_jitter=0.28)
```

```python
sns.regplot(x='size', y='total_bill',data=tips, ci=84,x_jitter=0.28,
            scatter_kws={'color':'red'})
```

```python
sns.regplot(x='size', y='total_bill',data=tips,x_estimator=np.mean,
            scatter_kws={'color':'green'})
```
