```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips= sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.boxplot(x=tips['size'])
```

```python
sns.boxplot(x=tips['total_bill'])
```

```python
sns.boxplot(x='sex',y='total_bill', data=tips,palette='autumn')
```

```python
sns.boxplot(x='day', y='total_bill',data=tips, palette='winter_r')
```

```python
sns.boxplot(x='day',y='total_bill', data=tips, hue='sex', palette='husl')
```

```python
sns.boxplot(x='day',y='total_bill', data=tips, hue='sex', palette='coolwarm')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips,hue='time')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, order=['Sun', 'Fri','Thur','Sat'], palette='spring')
```

```python
sns.boxplot(x='sex', y='total_bill', data=tips, order=['Female','Male'], palette='husl')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='husl')
sns.swarmplot(x='day',y='total_bill', data=tips,palette='autumn')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='husl')
sns.swarmplot(x='day',y='total_bill', data=tips,color='black')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='husl')
sns.swarmplot(x='day',y='total_bill', data=tips,color='0.45')
```
