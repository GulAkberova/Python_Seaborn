```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(iris, hue='species')
```

```python
sns.pairplot(iris,hue='species',palette='winter_r')
```

```python
sns.pairplot(iris,hue='species',palette='husl',markers=['o','D','s'])
```

```python
sns.pairplot(iris, vars=['sepal_length','sepal_width'],size=2)
```

```python
sns.pairplot(iris,x_vars=['petal_length', 'petal_width'],
             y_vars=['sepal_length', 'sepal_width'],
            palette='summer',hue='species')
```

```python
sns.pairplot(iris, diag_kind='kde',palette='spring',hue='species')
```

```python
sns.pairplot(iris,kind='reg')
```

```python
sns.pairplot(iris,diag_kind='kde',palette='autumn', hue='species',kind='reg')
```
