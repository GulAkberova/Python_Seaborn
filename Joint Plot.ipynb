```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
sns.jointplot(x='total_bill',y='tip', data=tips, kind='reg',color='green')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde',shade=True)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris, kind='kde',color='red'
             ,shade=True)
```

```python
from scipy.stats import spearmanr
```

```python
#sns.jointplot(x='total_bill', y='tip', data=tips, stat_func=spearmanr)
# Jointplot oluturun
g = sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg',color='Blue')

# Spearman korelasyon katsaysn hesaplayn ve grafie ekleyin
corr, _ = spearmanr(tips['total_bill'], tips['tip'])
g.ax_joint.text(0.05, 0.9, f'Spearman Correlation: {corr:.2f}', transform=g.ax_joint.transAxes)

plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, ratio=1, size=2)
```
