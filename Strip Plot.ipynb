```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.stripplot(x=tips['tip'],color='green')
```

```python
sns.stripplot(x=tips['total_bill'],color='red')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, palette='husl')
```

```python
sns.stripplot(x='day',y='tip',data=tips, jitter=False,palette='autumn')
```

```python
sns.stripplot(x='day',y='tip',data=tips, jitter=0.2,palette='autumn')
```

```python
sns.stripplot(x='total_bill', y='day', data=tips,palette='husl', jitter=0.2, linewidth=1.5)
```

```python
sns.stripplot(x='day',y='total_bill', data= tips, hue='sex',palette='spring')
```

```python
sns.stripplot(x='day',y='total_bill', data= tips, hue='sex',palette='winter_r',jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='smoker',jitter=True,
              palette='spring')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='smoker',jitter=True,
              palette='spring',order=['Fri','Sun','Sat','Thur'])
```

```python
sns.stripplot(x='sex',y='tip',data=tips,palette='autumn')
```

```python
sns.stripplot(x='sex',y='tip',data=tips,palette='husl',marker='D')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,palette='husl',marker='D',size=8)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,palette='autumn',marker='D',size=8,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,palette='husl',marker='D',size=8,hue='sex',
             edgecolor='black',alpha=0.3,dodge=True)
```

```python
sns.stripplot(x='tip',y='day',data=tips,palette='winter_r',color='1')
sns.boxplot(x='tip',y='day',data=tips,palette='spring')
```

```python
sns.stripplot(x='tip',y='day',data=tips,color='green')
sns.violinplot(x='tip',y='day',data=tips,color='0.9')
```
