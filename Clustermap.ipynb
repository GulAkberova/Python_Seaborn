```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights = flights.pivot(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(flights)
```

```python
sns.clustermap(flights)
```

```python
sns.clustermap(flights,col_cluster=False)
```

```python
sns.clustermap(flights,row_cluster=False)
```

```python
sns.clustermap(flights,cmap='coolwarm',linewidth=1)
```

```python
sns.clustermap(flights,cmap='Blues_r',linewidth=1.5)
```

```python
sns.clustermap(flights,cmap='coolwarm',linewidth=1.2,figsize=(6,5))
```

```python
sns.clustermap(flights,standard_scale=0)
```

```python
sns.clustermap(flights,z_score=1) #z_score 0 (rows) / 1 (columns)
```
